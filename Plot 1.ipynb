```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_excel (r'run1.xlsx')
plt.rcParams['font.size'] = '20'
x='Time'
y1='Data_1'
y2='Data_2'
df.plot(x, y= [y1, y2], kind='line', figsize=(10,5), color=['black','#FF0000'], label=['data 1','data 2'])
plt.title ("Kirsty's Plot", pad=20)
plt.xlabel('Time / ns')
plt.ylabel('Distance / nm')
plt.legend(loc='upper right', frameon=False)
plt.ylim([0.0, 10])
plt.xlim([0, 100])
plt.show
```
