```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel(r'spectrum.xlsx')
y='Spectrum'
x='Wavelength'
df.plot(x, y, kind='line',figsize=(10,5), fontsize=20, color=['red'], legend=False)
plt.title ("Absorption Spectrum", fontsize=30)
plt.xlabel('Wavelength / nm', fontsize=20)
plt.ylabel('Absorbance', fontsize=20)
plt.ylim([0.01, 1.05])
plt.xlim([350, 500])
plt.axvline(x=409,ymax=0.79,color='black',linestyle='--')
plt.show
```
